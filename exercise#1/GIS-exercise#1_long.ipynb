```python
import os

import pandas as pd
import geopandas as gpd
import numpy as np
from shapely.geometry import LineString
```

```python
os.chdir(r'C:\Users\BEIJING-c025\OneDrive\Materials\GIS\practice\practice_1')
```

```python
crs_init='epsg:4326'
crs_out='epsg:26718'
```

```python
df=pd.read_csv(r'202011-citibike-tripdata.csv')
df
```

```python
cols=df.columns.tolist()
cols
```

```python
start_cols=['start station latitude','start station longitude']
end_cols=['end station latitude','end station longitude']
```

```python
attr_cols=[i for i in cols if ((i in start_cols)==False)&((i in end_cols)==False)]
attr_cols
```

```python
df_attr=df[attr_cols]
df_attr
```

```python
df_start=df[start_cols]
df_start
```

```python
df_end=df[end_cols]
df_end
```

```python
startPoints=gpd.points_from_xy(df_start['start station longitude'],df_start['start station latitude'])
startPoints
```

```python
gdf_start=gpd.GeoDataFrame(df_start,geometry=startPoints,crs=crs_init)
gdf_start
```

```python
endPoints=gpd.points_from_xy(df_end['end station longitude'],df_end['end station latitude'])
endPoints
```

```python
gdf_end=gpd.GeoDataFrame(df_end,geometry=endPoints,crs=crs_init)
gdf_end
```

```python
N=df.shape[0]
lines=[LineString([gdf_start.geometry[i],gdf_end.geometry[i]]) for i in range(0,N)]
lines[0:10]
```

```python
df_tripline=pd.merge(df_start[start_cols],df_end[end_cols],left_index=True,right_index=True)
df_tripline
```

```python
gdf_tripline=gpd.GeoDataFrame(df_tripline,geometry=gpd.GeoSeries(lines),crs=crs_init)
gdf_tripline
```

```python
gdf_tripline.crs
```

```python
gdf_tripline=gdf_tripline.to_crs(crs_out)
gdf_tripline.crs
```

```python
gdf_tripline['linedistance']=gdf_tripline['geometry'].length
gdf_tripline
```

```python
gdf_biketrip=pd.merge(gdf_tripline,df_attr,left_index=True,right_index=True)
gdf_biketrip
```

```python
gdf_biketrip.rename(columns={"tripduration":"duration",
                            "start station id":"startid", 
                            "start station name":"startname",
                            "start station latitude":"start_lat",
                            "start station longitude":"start_lon",
                            "end station id":"endid",
                            "end station name":"endname",
                            "end station latitude":"end_lat",
                            "end station longitude":"end_lon",
                            "birth year":"birthyear",
                            "linedistance":"linedist"},inplace=True)

assert np.all(gdf_biketrip.columns.str.len() <= 10), \
"Column names will be truncated when saving as shapefile!"
```

```python
os.makedirs('output')
gdf_biketrip.to_file(os.path.join('output','biketrip.shp'))
```
